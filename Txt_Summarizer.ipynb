```python
from datasets import load_dataset
multinews = load_dataset("cnn_dailymail", '3.0.0', split = 'test')
```

```python
multinews.to_pandas()
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("t5-small")
```

```python
multi_news = multinews.train_test_split(test_size=0.2)
```

```python
prefix = 'summarize'
def process_function(example):
  inputs = [prefix + doc for doc in example['article']]
  model_inputs = tokenizer(inputs, max_length=1024, truncation=True)
  labels = tokenizer(text = example['highlights'], max_length=128, truncation=True)
  model_inputs['labels'] = labels['input_ids']
  return model_inputs
```

```python
tokenized_multi_news = multi_news.map(process_function, batched=True)
```

```python
from transformers import DataCollatorForSeq2Seq, AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model='t5-small')

model = AutoModelForSeq2SeqLM.from_pretrained('t5-small')
```

```python
trainings_args = Seq2SeqTrainingArguments(
    output_dir = './results',
    learning_rate = 2e-5,
    per_device_train_batch_size = 10,
    per_device_eval_batch_size = 10,
    weight_decay = 0.01, #regularization technique
    save_total_limit = 3,
    num_train_epochs = 10,
    fp16 = True
)
```

```python
trainer = Seq2SeqTrainer(
    model = model,
    args = trainings_args,
    train_dataset = tokenized_multi_news['train'],
    eval_dataset = tokenized_multi_news['test'],
    tokenizer = tokenizer,
    data_collator = data_collator
)
```

```python
trainer.train()
```
